```python
import numpy as np
import os
import matplotlib.pyplot as pyplot
from skimage.feature import greycomatrix, greycoprops
import pandas as pd
import cv2
from skimage.measure import label, regionprops

slices = []
proList = ['contrast', 'dissimilarity', 'homogeneity', 'ASM', 'energy']    # for GCLM texture analysis 
featlist = ['contrast', 'dissimilarity', 'homogeneity', 'ASM', 'energy', 'hue', 'value', 'saturaton', 'path', 'label']
properties = np.zeros(6) 
glcmMatrix = []
final = []
crop = "BPLD"  # Change this to your dataset name

# Define the classes and the number of images for each class
classes = {
    "Melanose": "Melanose",
    "healthy": "healthy",
    "greening": "greening",
    "canker": "canker",
    "Black spot": "Black spot"
}

# Update the folders list based on the classes
folders = list(classes.keys())

for folder in folders:
    print(folder) 
    labell = folders.index(folder)
    class_folder_path = f"D:/IDP/BPLD Dataset/{folder}/"  # Update the path accordingly

    image_folder_list = os.listdir(class_folder_path)
    # num_images = min(classes[folder], len(image_folder_list))  # Ensure we don't exceed the specified number of images

    for i in range(150):
        image_path = os.path.join(class_folder_path, image_folder_list[i])
        abc = cv2.imread(image_path)   # read image usingOpenCV
        gray_image = cv2.cvtColor(abc, cv2.COLOR_BGR2GRAY)  # converts RGB Image to gray image
        hsv = cv2.cvtColor(abc, cv2.COLOR_BGR2HSV)  # convert RGB image to HSV
        h, s, v = cv2.split(hsv)   # split hsv image to h , s , v
        h_mean = np.mean(h)  # calusculate Hue mean value
        s_mean = np.mean(s)
        v_mean = np.mean(v)
        
        glcmMatrix = greycomatrix(gray_image, [1], [0], levels=2 ** 8)   # computes GCLM of gray image
        for j in range(len(proList)):
            properties[j] = greycoprops(glcmMatrix, prop=proList[j])   # computes GCLM properties using greycoprops and stoes it in properties

        features = np.array(
            [properties[0], properties[1], properties[2], properties[3], properties[4], h_mean, s_mean, v_mean,
             image_path, labell])
        final.append(features)

df = pd.DataFrame(final, columns=featlist)
filepath = crop + "Final.xlsx"
df.to_excel(filepath)
```
